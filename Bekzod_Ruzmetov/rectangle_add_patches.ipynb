```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib inline
```

```python
rec1 = Rectangle((0.5, 0), 0.2, 0.8)
rec2 = Rectangle((0.2, 0), 0.2, 0.6)

fig, ax = plt.subplots()

ax.add_patch(rec1)
ax.add_patch(rec2)
```

```python
ax.patches[0].set_facecolor('red')

fig
```
