```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
np.random.seed(5)
x = np.linspace(1,40,10)

y1 = x/(np.random.rand(10)*2)
y2 = x/(np.random.rand(10)*3)

line1 = Line2D(x, y1, label='Son of a Line')
line2 = Line2D(x, y2, label='Another Line')

fig, ax = plt.subplots()

ax.add_line(line1)
ax.add_line(line2)

ax.autoscale()
```

```python
help(fig)
```

```python
fig.set_size_inches(15,5)
fig
```

```python
help(line1)
```

```python
l1 = ax.lines[0]
l2 = ax.lines[1]
```

```python
l1.set_alpha(0.3)
fig
```

```python
for line in ax.lines:
    line.set_color('red')
    line.set_lw(2)
fig
```

```python
l1.set_linestyle('dashed')
fig
```

```python
l2.set_marker('*')
l2.set_markersize(20)
l2.set_markerfacecolor('white')
l2.set_markevery(1)
l2.set_sketch_params(0,9)
fig
```

```python
l1.set_clip_on(False)
ax.set_xlim(0,39)

fig
```

```python
ax.set_title('Some line chart')
ax.set_xlabel('This is X axls')
ax.set_ylabel('This is Y axls')
fig
```

```python
fig.savefig('artist1.png', dpi=350, transparent=False)
```
